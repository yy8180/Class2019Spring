# data preparation and visualization

## 판다스 & Matplotlib

판다스 라이브러리를 이용해 csv파일을 읽어올 수 있다.

판다스 라이브러리는 데이터를 불러오고 전처리하는데 큰 도움을 준다.

또한 matplotlib라이브러리를 이용해 데이터를 시각화해 본다.

데이터를 살펴보기 위해 그래프로 나타내는 것이 중요할 때가 많다

```python
with open('./regression.txt',encoding='UTF8') as f:
    df=pd.read_csv(f)
    
X=df['X']
Y=df['Y']

df.head()
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X,Y)
plt.show()
```

# create model

```python
X.shape
```

Dense=기본적인 완전 연결 신경망
모델을 만들고, 컴파일을 해줘야 한다. 이때 오차 계산방법과 최적화 방법, 점수 계산 방식을 설정해준다.

여기서는 mse(mean squard error)와 sgd(Stochastic gradient descent)를 사용한다.

```python
model= Sequential()

model.add(Dense(1,input_shape=[1]))



model.compile(loss='mse',optimizer='sgd',metrics=['mae'])
```

# train model

모델의 예측값과 레이블의 원래 값을 비교해 가며 가중치를 업데이트한다.

```python
history=model.fit(X, Y,epochs=1000)
```

# model prediction

훈련이 완료된 모델을 이용해 주어진 x값에 대해 y값을 예측할 수 있다

```python
x=[x for x in range(0,11)]
y=model.predict(x)

plt.plot(X,Y, '.')
plt.plot(x,y,'rx')
```
